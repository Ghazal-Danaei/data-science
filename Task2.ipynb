Question 4:chi-2\
is there any relation between LMPD_DIVISION and CRIME_TYPE?\
H0:there is not any relation between LMPD_DIVISION and CRIME_TYPE.
H1:there is a relation between LMPD_DIVISION and CRIME_TYPE.

```python
data_crosstab = pd.crosstab(dataset['LMPD_DIVISION'], dataset['CRIME_TYPE'], margins = False)                          
#contingency table                                                           
print(data_crosstab)
```

```python
from scipy.stats import chi2_contingency

# Get chi-square value , p-value, degrees of freedom, expected frequencies using the function chi2_contingency
stat, p, dof, expected = chi2_contingency(data_crosstab)
# select significance value
alpha = 0.05
# Determine whether to reject or keep your null hypothesis
print('significance=%.3f, p=%.3f' % (alpha, p))
if p <= alpha:
    print('Variables are associated (reject H0)')
else:
    print('Variables are not associated(fail to reject H0)')
```
